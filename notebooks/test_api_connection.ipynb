#### Fetching historical data

```python
import os
from dotenv import load_dotenv
from alpaca.data.requests import StockQuotesRequest, StockBarsRequest
from alpaca.data import CryptoHistoricalDataClient, StockHistoricalDataClient, OptionHistoricalDataClient,TimeFrame
load_dotenv()



crypto_client = CryptoHistoricalDataClient()

# keys required
stock_client = StockHistoricalDataClient(os.getenv("ALPACA_KEY"),  os.getenv("ALPACA_SECRET"))
```

```python
request_params = StockBarsRequest(
    symbol_or_symbols=['AAPL'],
    timeframe=TimeFrame.Day,
    start="2024-01-01",
    )
```

```python
stock_client.get_stock_bars(request_params)
```

```python
from alpaca_trade_api.rest import REST, TimeFrame
api = REST(os.getenv("ALPACA_KEY"), os.getenv("ALPACA_SECRET"))

api.get_bars("AAPL", TimeFrame.Hour, "2021-06-08", "2021-06-08", adjustment='raw').df
```
